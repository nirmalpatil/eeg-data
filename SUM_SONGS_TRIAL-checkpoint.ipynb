```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import csv
```

```python
data = pd.read_csv(r"C:\Users\Nirmal\Desktop\FYP\FEATURE_TABLE - allthree.csv")
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
sns.pairplot(df,hue='State',palette = 'coolwarm')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['State','T7_var','F7_var','P3_var','FPZ_var','F8_var','T8_var','P4_var','T7_avg','F7_avg','P3_avg','FPZ_avg','F8_avg','T8_avg','P4_avg',
'T7_stdev','F7_stdev','P3_stdev','FPZ_stdev','F8_stdev','T8_stdev','P4_stdev'



           ],axis=1)
y = df['State']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
from sklearn.svm import SVC
```

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train,y_train)
```

```python
predictions = svc_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```

```python
svc_model.decision_function(X)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C': [0.1,1, 10, 100], 'gamma': [1,0.1,0.01,0.001]} 
```

```python
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=2)
grid.fit(X_train,y_train)
```

```python
grid_predictions = grid.predict(X_test)
```

```python
print(confusion_matrix(y_test,grid_predictions))
```

```python
print(classification_report(y_test,grid_predictions))
```
